```python
import pandas as pd
df= pd.read_csv('~/PHYS476/hw1_keras/breast-cancer-wisconsin.data')
df.columns = ['id', 'clump', 'cell_size', 'cell_shape', 'adhesion', 'epithelial',
                'bare_nuclei', 'chromatin', 'nucleoli', 'mitoses', 'class']
df = df.apply(pd.to_numeric, errors='coerce')
df.dropna(axis=0, how='any', inplace=True)
df = pd.get_dummies(df, columns=['class'], prefix='class')
print(df)
```

```python
from sklearn.model_selection import train_test_split
labels = df[['class_2', 'class_4']]
df.drop(['id', 'class_2', 'class_4'], axis=1, inplace= True)
print(df)
print(labels)
x_train, x_test, y_train, y_test = train_test_split(df, labels, test_size=0.1)
```

```python
import tensorflow as tf
# Parameters
learning_rate = 0.1
num_epoch = 200
batch_size = 32
display_step = 100

# Network Parameters
n_input = 9 # size of input
n_layer1 = 12 # 1st layer number of neurons
n_layer2 = 7 # 2nd layer number of neurons
# n_layer3 = 5
# n_layer4 = 3
n_output = 2 # size of output: 2 classes

# Create place holder for input and output layers
x = tf.placeholder(tf.float32, [None, n_input])
y = tf.placeholder(tf.float32, [None, n_output])
```

```python
# Store layers weight & bias
weights = {
    'l1': tf.Variable(tf.random_normal([n_input, n_layer1])),
    'l2': tf.Variable(tf.random_normal([n_layer1, n_layer2])),
#     'l3': tf.Variable(tf.random_normal([n_layer2, n_layer3])),
#     'l4': tf.Variable(tf.random_normal([n_layer3, n_layer4])),
    'out': tf.Variable(tf.random_normal([n_layer2, n_output]))
}
biases = {
    'b1': tf.Variable(tf.random_normal([n_layer1])),
    'b2': tf.Variable(tf.random_normal([n_layer2])),
    'b3': tf.Variable(tf.random_normal([n_layer3])),
    'b4': tf.Variable(tf.random_normal([n_layer4])),
    'out': tf.Variable(tf.random_normal([n_output]))
}
# temp = tf.Variable(tf.random_normal([1, n_input]))
# with tf.Session() as sess:
#     sess.run(tf.global_variables_initializer())
#     result = tf.matmul(temp, weights['l1'])
#     print(temp, weights['l1'])
#     print(result)
#     print(sess.run(result))
```

```python
# Create deep NN model
def nn_model(x):
    # linear algebra matrix * vector + bias vector
    layer1 = tf.add(tf.matmul(x, weights['l1']), biases['b1'])
    layer1 = tf.nn.relu(layer1)
    layer2 = tf.add(tf.matmul(layer1, weights['l2']), biases['b2'])
    layer2 = tf.nn.relu(layer2)
#     layer3 = tf.add(tf.matmul(layer2, weights['l3']), biases['b3'])
#     layer3 = tf.nn.relu(layer3)
#     layer4 = tf.add(tf.matmul(layer3, weights['l4']), biases['b4'])
#     layer4 = tf.nn.relu(layer4)
    out_layer = tf.add(tf.matmul(layer2, weights['out']), biases['out'])
    return out_layer
```

```python
# construct model
logits = nn_model(x)

# define loss function and optimizer
# Computes softmax cross entropy between logits and labels.
loss_func = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=y))
op = tf.train.RMSPropOptimizer(learning_rate=learning_rate)
train_op = op.minimize(loss_func)

# Evaluate model 
# assume the output layer is one-hot encoded, the index of max value(i.e. 1) is the correct category
acc_pred = tf.equal(tf.argmax(input=logits, axis=1), tf.argmax(input=y, axis=1))
# cast to float because reduce_mean has strong type dependent on input type
accuracy = tf.reduce_mean(tf.cast(acc_pred, dtype=tf.float32))
```

```python
from sklearn.utils import shuffle

# Training
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for e in range(0, num_epoch):
        # need something to split into batches
        batch_x, batch_y = shuffle(x_train, y_train, n_samples=batch_size)
        sess.run(train_op, feed_dict={x: batch_x, y: batch_y})
        if e % display_step == 0 or e == num_epoch-1:
            # Calculate batch loss and accuracy
            loss, acc = sess.run([loss_func, accuracy], feed_dict={x: batch_x,
                                                                 y: batch_y})
            print("Step " + str(e) + ", Minibatch Loss= " + \
                  "{:.4f}".format(loss) + ", Training Accuracy= " + \
                  "{:.3f}".format(acc))
    print("Optimization Finished!")

    # Calculate accuracy for test data
    print('Test Accuracy',  sess.run(accuracy, feed_dict={x: x_test, y: y_test}))
```
